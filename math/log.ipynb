```python
import math
import torch
import matplotlib.pyplot as plt
```

```python
x1 = []
x2 = []
x3 = []
x4 = []
x5 = []
x6 = []
```

```python
for i in range(100):
    if (i > 1):
        x1.append(math.log(i))
        x2.append(math.log2(i))
        x3.append(math.log10(i))
        x4.append(math.log(math.log(i)))
        x5.append(math.log2(math.log2(i)))
        x6.append(math.log10(math.log10(i)))

        print(i, math.log(i), math.log2(i), math.log10(i), math.log(math.log(i)), math.log2(math.log2(i)), math.log10(math.log10(i)))
```

```python
plt.subplot(2, 3, 1)
plt.plot(x1)
plt.hlines(y=3.9, xmin=0.0, xmax=100.0, color='r')
plt.vlines(x=50, ymin=0.0, ymax=5.0, color='r')
plt.subplot(2, 3, 2)
plt.plot(x2)
plt.subplot(2, 3, 3)
plt.plot(x3)
plt.subplot(2, 3, 4)
plt.plot(x4)
plt.subplot(2, 3, 5)
plt.plot(x5)
plt.subplot(2, 3, 6)
plt.plot(x6)
```
