```python
scenario1 = r'//' # 0,2 - index 0 - 2
scenario2 = r'//*\n' #2,3 - index 2 do 3 --> start dlugiego komentarza (bo sa jeszcze te pomiedzy liniami i na koncu lini kodu)
scenario3 = r' \*' # kontynuacja dlugiego komentarza pomiedzy funkcjami na poczatku lini
scenario4 = r' \*/\n' # koniec dlugiego komentarza - ALE NIE SPRAWDZAJ TEGO ZE SOBA TYLKO WYCINAJ WSZYSTKO CO PASUJE !!!
scenario5 = r'#include' # 0,8 - wazne sa te indeksy ! i to ile jest powtorzen !!
```

```python
#https://pynative.com/python-find-position-of-regex-match-using-span-start-end/
cnt = 0
pattern = r'#include'
#test_string = "// SPDX-License-Identifier: GPL-2.0-only"
test_string = test_l
for match in re.finditer(pattern, test_string, re.IGNORECASE):
    cnt += 1
    print( cnt, " st match start index", match.start(), " end match index " , match.end())
```

```python
#https://pynative.com/python-find-position-of-regex-match-using-span-start-end/
cnt = 0
pattern_number = 0
#pattern = r'#include'
#test_string = "// SPDX-License-Identifier: GPL-2.0-only"
test_string = test_l
for pattern in [scenario1,scenario2,scenario3,scenario4,scenario5]:
    for match in re.finditer(pattern, test_string, re.IGNORECASE):
        cnt += 1
        print( pattern, " [ ", pattern_number, " ] " , cnt, " st match start index", match.start(), " end match index " , match.end())
    pattern_number += 1
```

```python
#https://pynative.com/python-find-position-of-regex-match-using-span-start-end/
cnt = 0
pattern_number = 0
line_number = 0;
skipped = 0
#pattern = r'#include'
#test_string = "// SPDX-License-Identifier: GPL-2.0-only"
#test_string = test_l
matched_lines = []
for test_string in lines[:]:
    pattern_number = 0 # reset for each line
    for pattern in [scenario1,scenario2,scenario3,scenario4,scenario5]:
        for match in re.finditer(pattern, test_string, re.IGNORECASE):
            # cut scenario1 and scenario2
            if match.start() >= 0 and match.end() <=4:
                skipped += 1
                #continue
            cnt += 1
            # scenario cut - NOT MATCH
            if pattern_number == 2 and match.end() > 4:
                # NOT MATCH
                continue
            if pattern_number == 1 and match.end() > 4:
                # NOT MATCH
                continue
            if pattern_number == 3 and match.end() > 4:
                # NOT MATCH
                continue
            print( "( ", line_number, " ) ", pattern, " [ ", pattern_number, " ] " , cnt, " st match start index", match.start(), 
                  " end match index " , match.end(), " skipped: ", skipped, " line ", test_string)
            lines[line_number] = "\n"
            matched_lines.append(test_string)
            
        pattern_number += 1
    line_number += 1
```
